```python
#aiter
#return asynchronous iterator
#비동기 반복자를 반환하는 함수
async def async_generator():
    for i in range(3):
        yield i


iterator = aiter(async_generator())

async def main():
    async for value in iterator:
        print(value)
```

```python
#anext
#return next item from asynchronous iterator
#비동기 반복자에서 다음 값을 가져오는 함수
async def async_generator():
    for i in range(3):
        yield i

async def main():
    iterator = aiter(async_generator())

    value = await anext(iterator)
    print(value)
```

```python
#classmethod
#convert method to class method
#메서드를 클래스 메서드로 만드는 데코레이터
class Person:
    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

print(Person.get_count())
p1 = Person()
p2 = Person()
print(Person.get_count())
```

```python
#super
#access parent class methods
#부모 클래스의 메서드에 접근하는 함수
class A:
    def hello(self):
        print("A")

class B(A):
    def hello(self):
        super().hello()
        print("B")

b = B()
b.hello()
```
